```python
merged_bank_customer=bank_additional.merge(customer_details_2012_2014,how='inner',left_on='id_',right_on='ID')
merged_bank_customer.drop('id_',axis=1,inplace=True)
merged_bank_customer.head()
```

```python
merged_bank_customer.columns
```

```python
columnas_renombradas={columna:columna.lower().replace('.','_') for columna in merged_bank_customer.columns}
columnas_renombradas
```

```python
merged_bank_customer.rename(columns=columnas_renombradas, inplace=True)
merged_bank_customer.head()
```

```python
for columna in merged_bank_customer.columns.to_list():
  try:
    merged_bank_customer[columna]=merged_bank_customer[columna].str.lower().replace('.','_')
  except:
    pass
merged_bank_customer.head()
```

```python
fecha= merged_bank_customer['date'].str.split('-', expand=True) #expand sirve para generar columnas para cada uno de los elementos separados mediante split. Esas columnas se agrupan en un dataframe que, además, tiene una columna como índice. 
fecha #esto es un dataframe
```

```python
merged_bank_customer['contact_month']=fecha[1]
merged_bank_customer['contact_year']=fecha[2]
merged_bank_customer.head()
```

```python
meses = {
    "enero": '01',
    "febrero": '02',
    "marzo": '03',
    "abril": '04',
    "mayo": '05',
    "junio": '06',
    "julio": '07',
    "agosto": '08',
    "septiembre": '09',
    "octubre": '10',
    "noviembre": '11',
    "diciembre": '12'
}
```

```python
merged_bank_customer.replace({'date': meses},regex=True,inplace=True)
merged_bank_customer.head()
```

```python
merged_bank_customer['date']=pd.to_datetime(merged_bank_customer['date'],format='%d-%m-%Y')
```

```python
merged_bank_customer.info()
```

```python
merged_bank_customer.head()
```

```python
merged_bank_customer['education']=merged_bank_customer['education'].str.replace('.',' ')
merged_bank_customer.head()
```

```python
merged_bank_customer.columns
```

```python
columnas_str_float=['cons_price_idx','cons_conf_idx','euribor3m','nr_employed']
for columna in columnas_str_float:
  merged_bank_customer[columna]=merged_bank_customer[columna].str.replace(',','.').astype(float)
```

```python
merged_bank_customer.head()
```

```python
merged_bank_customer.info()
```

```python
diccionario_boolanos={0:'no',1:'si'}
lista_col_booleanos={'default','housing','loan'}

for columna in lista_col_booleanos:
  merged_bank_customer[columna]=merged_bank_customer[columna].map(diccionario_boolanos)

#Otra opción: bank[lista_col_bool] = bank[lista_col_bool].map(lambda x:diccionario_booleanos.get(x))
```

```python
merged_bank_customer['pdays']=merged_bank_customer['pdays'].replace(999,np.nan)
```
