```python
from scipy import stats
result=stats.shapiro(df['rm'])
result.statistic
result.pvalue
print('Test statistic:',result.statistic)
print('P-value:','{:e}'.format(result.pvalue))
```

```python
result=stats.kstest(df['rm'],'norm')
result.statistic
result.pvalue
print('D statistic:',result.statistic)
print('P-value:','{:e}'.format(result.pvalue))
```

```python
result=stats.anderson(df['rm'],'norm')
print("Critical Values:", result.critical_values)
print("Anderson-Darling Test Statistic:", result.statistic)
print("Significance Levels:", result.significance_level)
```

```python
stats.probplot(df['rm'], dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```
