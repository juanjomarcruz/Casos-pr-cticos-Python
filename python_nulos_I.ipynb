```python
# Tratamiento de datos
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

# Imputación de nulos
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

# Librerías de visualización
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
bank_limpio = pd.read_csv('bank-limpio.csv', index_col=0)
bank_limpio.head()
```

```python
bank_limpio.info()
```

```python
bank_limpio.columns
```

```python
nulos=round(bank_limpio.isnull().sum()/bank_limpio.shape[0]*100)
nulos
```

```python
#para que no nos salgan todas las columnas, ya que queremos solo las que tienen nulos, vamos a crear un dataframe convirtiendo la serie de nulos y después filtrando ese df por aquellos en los que el porcentaje de nulos es mayor que 0.

df_nulos=pd.DataFrame(nulos,columns=['%_nulos'])
df_nulos
```

```python
df_nulos[df_nulos['%_nulos']>0]
```

```python
#seleccionamos las columnas categóricas del dataframe que tienen algún nulo y son de tipo object (categóricas).
nul_cat_columns=bank_limpio[bank_limpio.columns[bank_limpio.isnull().any()]].select_dtypes('object').columns
```

```python
for columna in nul_cat_columns:
  print(f'La distribución de la categorías para la columna {columna.upper()}:')
  display(round(bank_limpio[columna].value_counts()/bank_limpio.shape[0]*100))
  print("........................")
```

```python
#¿Qué haremos para gestionar esos nulos? Tenemos dos opciones: sustituirlos por unknown o por la moda, es decir, por el valor que más se repite. Si no tenemos un valor que se repita claramente, lo sustituiremos por una categoría nueva 'Unknown'.

fill_by_mode=['marital', 'loan','default']
fill_by_unknown=['job','education','housing','contact_month','date']
```

```python
for columna in fill_by_mode:
  moda=bank_limpio[columna].mode()[0] #cogemos el primer elemento
  bank_limpio[columna]=bank_limpio[columna].fillna(moda)
print("Después del reemplazo usando 'fillna' quedan los siguientes nulos")
bank_limpio[fill_by_mode].isnull().sum()
```

```python
for columna in fill_by_unknown:
  bank_limpio[columna]=bank_limpio[columna].fillna('unknown')
print("Después del reemplazo usando 'fillna' quedan los siguientes nulos")
bank_limpio[fill_by_unknown].isnull().sum()
```

```python
bank_limpio.info()
```

```python
#obtenemos una lista de las columnas numéricas con algún valor nulo.
nul_num_columns=bank_limpio[bank_limpio.columns[bank_limpio.isnull().any()]].select_dtypes(include=['float64','int64']).columns
```

```python
round(bank_limpio[nul_num_columns].isnull().sum()/bank_limpio.shape[0]*100)
```

Antes de continuar con la gestión de nulos de las columnas numéricas echemos un vistazo a las columnas que tenemos. ¿Son todas las columnas numéricas o hay alguna falsa numérica, es decir, que tengamos que tratar como categórica? 

La respuesta es sí. La columna contact_year, ya que aunque sea un número es un dato tipo fecha, no tiene sentido tratarlo como un número ya que no podríamos sustituir sus nulos con, por ejemplo, la media. 

Así que vamos a gestionar sus nulos antes de nada. Como tiene muy pocos los sustituiremos por el valor más frecuente(moda). Para ello tendremos que imputar con el SimpleImputer.(Hace lo mismo que el método.fillna())

```python
# Primero creamos el objeto de la clase del SimpleImputer
imputer_contact_year = SimpleImputer(strategy = "most_frequent")# Usamos el 'most_frequent' ya que queremos cambiarlo por la moda

bank_limpio["contact_year"] = imputer_contact_year.fit_transform(bank_limpio[["contact_year"]])

# Por último comprobamos si tenemos nulos. 
print(f"Después del 'SimpleImputer' tenemos {bank_limpio['contact_year'].isnull().sum()} nulos")
```

```python
#La columna numérica pdays tiene un alto porcentaje de nulos (96%). Los sustituiremos por la categoría unknown.

bank_limpio['pdays']=bank_limpio['pdays'].fillna('unknown')
```

```python
round(bank_limpio[nul_num_columns].isnull().sum()/bank_limpio.shape[0]*100)
```

```python
fig = make_subplots(
    rows = 2, cols = 2,
    specs = [[{"colspan": 2},None], [{"type": "xy"},{"type": "xy"}]], subplot_titles = ['age', 'cons_price_idx','euribor3m'])
fig.add_trace(go.Box(x = bank_limpio['age']
             ),
             row = 1,col = 1)
fig.add_trace(go.Box(x = bank_limpio['cons_price_idx']
             ),
             row = 2,col = 1),
fig.add_trace(go.Box(x = bank_limpio['euribor3m']
             ),
             row = 2,col = 2)
```

Como podemos observar,la columna 'age' posee outliers, además, la columna 'euribor3m' posee una cantidad elevada de nulos (21%), por lo tanto para estas columnas usaremos técnicas de gestión de nulos más avanzadas.

Por otro lado, la columna 'cons_price_idx' posee muy pocos nulos y no tiene outliers, por lo tanto podemos reemplazar sus nulos por la mediana o por la media.

```python
media=bank_limpio['cons_price_idx'].mean()
mediana=bank_limpio['cons_price_idx'].median()
df_col_cons_price=pd.DataFrame(bank_limpio['cons_price_idx'])
df_col_cons_price['substituted_by_mean']=df_col_cons_price['cons_price_idx'].fillna(media)
df_col_cons_price['substituted_by_median']=df_col_cons_price['cons_price_idx'].fillna(mediana)
display(df_col_cons_price['substituted_by_mean'].describe()['std'])
display(df_col_cons_price['substituted_by_median'].describe()['std'])
```

La desviación es menor en el caso de la media, por lo que en nuestro DataFrame original sustituiremos los nulos por ésta.

```python
bank_limpio['cons_price_idx']=bank_limpio['cons_price_idx'].fillna(media)
print(f'La columna tratada se queda con {bank_limpio['cons_price_idx'].isnull().sum()} valores nulos')
```

```python
round(bank_limpio[nul_num_columns].isnull().sum()/bank_limpio.shape[0]*100)
```

```python
#creamos dataframe con columnas a imputar nulos.
columnas_a_imputar=pd.DataFrame(bank_limpio[['age','euribor3m']])
#Creamos el objeto imputador_iterativo de la classe IterativeImputer. Este imputador iterativo es capaz de rellenar los valores nulos de una columna haciendo una predicción a partir de todos los valores del resto de columnas y de los valores no nulos de esa misma columna. El imputador actúa sobre cada colummna del dataframe para poder rellenar todos los valores nulos y hace tantas iteraciones como programemos. Se aplica un modelo de machine learning (por defecto, un modelo de regresión lineal) que es entrenado con los valores del resto de columnas y los valores no nulos de la columna (inputs) para generar el output (rellenar alor nulo de la columna en cuestión).
Imputador_iterativo=IterativeImputer(max_iter=20,random_state=42)
# Ajustamos y tranformamos los datos
imputer_iterative_imputado = Imputador_iterativo.fit_transform(columnas_a_imputar)
# Comprobamos que es lo que nos devuelve, que en este caso es un array
imputer_iterative_imputado
```

```python
# En este caso como vamos a aplicar los dos métodos sobre estas columnas, así que vamos a crear dos columnas nuevas
columnas_a_imputar[["age_iterative", "euribor_iterative"]] = imputer_iterative_imputado

# Comprobamos los nulos
print(f"Después del 'Iterative' tenemos: \n{columnas_a_imputar[['age_iterative', 'euribor_iterative']].isnull().sum()} nulos")
```

```python
Imputador_KNN=KNNImputer(n_neighbors=5)
imputado_KNN=Imputador_KNN.fit_transform(columnas_a_imputar[['age','euribor3m']])
imputado_KNN
```

```python
columnas_a_imputar[['age_knn','euribor_knn']]=imputado_KNN
# Comprobamos los nulos
print(f"Después del 'KNN' tenemos: \n{columnas_a_imputar[['age_knn', 'euribor_knn']].isnull().sum()} nulos")
columnas_a_imputar.describe().T
```

Ahora es momento de ir columna por columna para ver si el iterative o el knn es más correcto.
- **age**: En este caso la media es más parecida la del iterative, además la desviación estándar es menor. Por otro lado, la mediana es más parecida en el caso del KNN. Nos vamos a quedar con este último ya que la mediana del iterative se distanciaba casi dos unidades del original.

- **euribor3m**: En este caso, la media es más parecida la del iterative y la desviación es menor, pero la mediana es más parecida la del KNN. Al igual que en el caso anterior, las diferencias son menores en el knn.

```python
bank_limpio['age']=columnas_a_imputar['age_knn']
bank_limpio['euribor3m']=columnas_a_imputar['euribor_knn']
bank_limpio.info()
```
