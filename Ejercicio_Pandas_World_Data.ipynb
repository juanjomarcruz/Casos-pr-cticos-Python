```python
#3.4.1. Crea una función para añadir una nueva columna llamada 'Percent Urban Population' al DataFrame para indicar a qué porcentaje de población urbana tiene cada país.

def Percent_Urban_Population(row, density_col, land_area_col, urban_population_col):
    try:
        density = row[density_col]
        land_area = row[land_area_col]
        urban_population = row[urban_population_col]
        population = density * land_area
        percent_urban_population = round(urban_population / population * 100,2)
        return percent_urban_population
    except:
        np.nan

#Esta versión de la función Percent_Urban_Population toma cuatro argumentos: row (la fila del DataFrame), density_col, land_area_col y urban_population_col (los nombres de las columnas que contienen densidad, área de tierra y población urbana, respectivamente). Esto te permite reutilizar la función con diferentes DataFrames que tengan columnas con nombres diferentes pero contengan información similar.

world['Percent_urban_population'] = world.apply(Percent_Urban_Population, axis=1, args=('Density\n(P/Km2)', 'Land Area(Km2)', 'Urban_population'))
```

```python
#3.5.1. ¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:
#- Primero, necesitas calcular la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna "total_tax_rate".
#- Luego, calcula la densidad de población para cada continente. La densidad de población se encuentra en la columna "density".
#- Ahora, divide la tasa de impuestos total entre la densidad de población para obtener la tasa de impuestos total por persona por kilómetro cuadrado.
#- Encuentra el país que tiene el valor más alto en esta nueva columna calculada.

world['Total tax rate per person per Km2']= world['Total tax rate']/world['Density\n(P/Km2)']
world.groupby(world['Continent'])['Total tax rate per person per Km2'].max().reset_index(name='Maximum Total tax rate per person per Km2')
```

```python
#3.5.2. ¿Cuál es la diferencia entre el PIB promedio y el gasto en salud de bolsillo promedio por idioma oficial? Para contestar esta pregunta dabes seguir los siguientes pasos:
#- Agrupa el DataFrame por idioma oficial y calcula la media del Gasto de Bolsillo en salud (out_of_pocket_health_expenditure) y del GPD.
#- Calcula la diferencia entre el PIB promedio y el gasto en salud de bolsillo promedio

mean_GDP_by_language=world.groupby('Official language')['GDP'].mean()
mean_OOPHE_by_language=world.groupby('Official language')['Out of pocket health expenditure'].mean()
difference=mean_GDP_by_language-mean_OOPHE_by_language
difference.reset_index(name='Mean GDP - Mean OOP Health expenditure')
```

```python
#3.5.3. ¿Cuál es la tasa de maternidad promedio por tasa de impuestos total? Para resolver el ejercicio deberás:
#- Primero, calcula la tasa de maternidad promedio para cada continente. Puedes usar la columna "maternal_mortality_ratio" que representa la tasa de mortalidad materna.
#- Calcula la tasa de impuestos total promedio para cada continente. Puedes usar la columna "total_tax_rate" que representa la tasa de impuestos total en cada continente.
#- Luego, divide la tasa de maternidad promedio por la tasa de impuestos total promedio para obtener la relación entre ambas tasas.
#Encuentra el país con la tasa de maternidad más alta por punto porcentual de tasa de impuestos total.

mean_maternal_mortality_ratio_by_continent=world.groupby(world['Continent'])['Maternal mortality ratio'].mean()
mean_total_tax_rate_by_continent=world.groupby(world['Continent'])['Total tax rate'].mean()
relation=mean_maternal_mortality_ratio_by_continent/mean_total_tax_rate_by_continent
relation.reset_index(name='Mean maternal mortality per total tax rate ratio').max()
```

```python
#3.5.5. ¿Cuál es la diferencia entre la tasa de impuestos total y la tasa de desempleo promedio por continente?

Mean_total_tax_rate_by_continent=world.groupby(world['Continent'])['Total tax rate'].mean()
Mean_unemployment_rate_by_continent=world.groupby(world['Continent'])['Unemployment rate'].mean()
Difference=Mean_total_tax_rate_by_continent-Mean_unemployment_rate_by_continent
Difference.reset_index(name='Diferencia entre la tasa de impuestos total y la tasa de desempleo promedio')
```
